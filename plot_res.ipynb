```python
import os
import sys
import glob
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
from pathlib import Path
import matplotlib.pyplot as plt

from os.path import join

%matplotlib inline
```

```python
output_dir = './outputs'
```

### Pancreas

```python
dt = 'Pancreas'
ri = 1

fdn = os.listdir(join(output_dir, dt))[0]
dt_dir = join(output_dir, dt, fdn, f'results{ri}')

ad_pancreas = sc.read_h5ad(join(dt_dir, 'ad_120.h5ad'))
sc.pl.umap(ad_pancreas, color=['batchlb', 'CellType'])
```
